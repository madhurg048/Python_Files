**Method-1**
 
- Whenever we read the file

    - file
 
    - mode
 
        - by default it is 'r' = read mode
     
        - if you want to write it will apply 'w'
 
    - encode

        - encoding is used to avoid spl charachters
     
        - by default cp1252
     
        - if you want to avoid spl charchters = '\utf-8-sig'

```python
file_path = r"C:\Users\madhu\Documents\Naresh IT\Data files\mbox-short.txt"
file = open(file_path,
           mode = 'r',
           encoding = 'utf-8-sig')
print(file.read())
```

**Enumerate**

- Enumerate will give two values

- One is index and another one is line

- enumerate is similar to zip

```python
# Q3) Without counter find the number of lines
file_path="mbox-short.txt"
file=open(file_path,mode='r',encoding='utf-8-sig')
for index,line in enumerate(file):
    print(index,line)
```
